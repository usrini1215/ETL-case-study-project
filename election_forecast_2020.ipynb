```python
import pandas as pd
import numpy as np
```

```python
# convert csv files to df: probabilities of candidates winning electoral votes 
pres_ev_df = pd.read_csv("Data/presidential_ev_probabilities_2020.csv")
pres_ev_df
```

```python
# convert csv files to df: presidential poll averages 2020
pres_poll_df = pd.read_csv("Data/presidential_poll_averages_2020.csv")
```

```python
pres_poll_df
```

```python
# dropping unneeded columns
pres_ev_df=pres_ev_df.drop(columns=['cycle', 'branch','model','candidate_3rd','evprob_3rd','simulations', 'timestamp'])
```

```python
# rename columns
pres_ev_df=pres_ev_df.rename(columns={"modeldate": "Simulations Model Date", "candidate_inc": "Incumbent", "candidate_chal": "Challenger","evprob_inc":"EV Probability Incumbent",
                           "evprob_chal": "EV Probability Challenger", "total_ev": "Total EV"})
```

```python
pres_ev_df
```

```python
pres_poll_df=pres_poll_df.drop(columns=['cycle', 'pct_estimate'])
pres_poll_df
```

```python
pres_poll_df=pres_poll_df.rename(columns={"state": "State","modeldate": "Simulations Model Date", "candidate_name": "Candidate Name", "pct_trend_adjusted": "Poll Percentage"})
```

```python
pres_poll_df
```

```python
pres_poll_df["Candidate Name"]=pres_poll_df['Candidate Name'].replace(['Donald Trump', 'Joseph R. Biden Jr.' ], ["Trump", "Biden"])
```
